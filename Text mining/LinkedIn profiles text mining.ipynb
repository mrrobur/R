## Objective
Dataset used for this analysis are scraped LinkedIn profile pages of 30 professionals working in business, start-up and e-commerce area. The goal is to perform text mining, find frequent terms and determine similarities.

```r
rm(list = ls())
library(tm)
library(SnowballC)
```

```r
data_source = DirSource("txt")

docs<-VCorpus(data_source)
doc_num<-30 

docs<-tm_map(docs,removePunctuation)
docs<-tm_map(docs,removeNumbers)
docs<-tm_map(docs,removeWords,stopwords("english"))
docs<-tm_map(docs,tolower)
docs<-tm_map(docs,removeWords,tolower(c("Summary", "Experience", "Page", "Projects", "Languages", "Education", "year","years", "month","months","polish","poland")))
docs<-tm_map(docs,removeWords,c("linkedin","contact","skills","the",tolower(month.name),format(ISOdate(2004,1:12,1),"%B")))

docs<-tm_map(docs, stripWhitespace)
docs<-tm_map(docs, PlainTextDocument)
#docs<-tm_map(docs,stemDocument)
```

Scraped text files are loaded to the script and later on range of text mining cleansing operations are performed:
- removing punctuation
- removing digits
- removing stopwords
- removing list of defined words to be excluded
- changing all letters into lowercase
- stripping white space characters
Word stemming is not being used as in the most common words in CVs that will be analyzed do not usually occur in multipple variations

```r
dtm<-DocumentTermMatrix(docs)
dtm<-removeSparseTerms(dtm,0.40)

freq<-colSums(as.matrix(dtm))

ord<-sort(freq,decreasing=TRUE)
```

I create document terms matrix and remove sparse terms leaving out 40% of most common phrases

```r
#plot wordcloud ------------------
library(wordcloud)
set.seed(25)
#options(repr.plot.width=7, repr.plot.height=7)
wordcloud(words = names(ord),freq = ord, min.freq = 15, colors=brewer.pal(8, "Dark2"))
```

The wordcloud above presents the most frequent words in CV documents. The larger the caption appears the more frequent partivular word appears. Most common words are 'business' and 'marketing'

## Correlation
Next step is calculating cosine similarity of documents pairwise and plotting on a chart

```r
#cosine similarity-----------------------
#For cosine measure I use raw data, not Inverse Frequency
cos<-matrix(,doc_num,doc_num)
for(i in seq(1,doc_num)){
  for(j in seq(i,doc_num))
  {
    sum_h<-sum(as.matrix(dtm[i,]*dtm[j,]))
    sq_hx<-sqrt(sum(as.matrix(dtm[i,])^2))
    sq_hy<-sqrt(sum(as.matrix(dtm[j,])^2))
    cos[i,j]<-sum_h/(sq_hx*sq_hy)
   # cos[i,i]<-NA
  }
}

library('corrplot')
plot_matrix = matrix(,doc_num,doc_num)
#plotting cosinus similarity ratio ---------------
for(i in seq(1,doc_num)){
  for(j in seq(1,doc_num)){
    if(i<j){
      plot_matrix[i,j] = cos[i,j]
    }else{
      plot_matrix[i,j] = cos[j,i]
    }
  }
}
corrplot(plot_matrix, method="color", type="upper", tl.col = "black", tl.srt = 90, is.corr=FALSE, cl.lim=c(0,1.00001))
```

As correlation plot discards, some documents are very likewise, highest pair got 89% similarity measure. We can conclude that those two person have very similar area of expertise. Improving the model possibly might could let HR specialist perform look-a-like search for a perfect candidate based on a pattern. 
Particularly in our case the highest score is a false result. As it turns out the result was strongly influenced by the fact that those two person have the same name and plenty of recommendation where their name appears.
Second high score 88% though matched two independent entreprenuers having large input into startup companies and occupaing CEO roles currently and in the past.

The lowest score is around 10%, it compared a marketing specialist with a descriptive resume with an IT specialist who focus on hard skills and technologies on his profile.

## Clustering text
Lastly, I want to attemp clustering on scraped words to see whether they can be grouped into meaningful groups by algorithm

```r
#clustering--------
library(fpc)
library(cluster)

#docss<-tm_map(docs,stemDocument)
dtmss<-DocumentTermMatrix(docs)
dtmss<-removeSparseTerms(dtmss,0.4)

d<-dist(t(dtmss),method="euclidian")

cluster_num<-5
kfit<-kmeans(d,cluster_num)
for(i in seq(1,cluster_num))
{
  cat("Cluster ", i, ":", dtmss$dimnames$Terms[kfit$cluster==i],"\n\n")
}
```

Clustering algorithm splitted the group using eculidean metric. 
- Clsuter 1, base business words usualy co-existing
- Cluster 2, mainly words linked to startup environment and technology
- Cluster 3, frequent words link
- Cluster 4, most common words
- Cluster 5, words focused on people, team and general management
